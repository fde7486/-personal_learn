# covid19 資料探索

## 讀取資料

```python
df = pd.read_csv('covid19_data/time_series_covid19_confirmed_global.csv')
df.head(10)
```

## 將日期轉換成 DatetimeIndex

```python
# 只有少數國家有 Province/State，所以我們把 Province/State 捨去，以 Country/Region 為單位來看確診數
df = df.drop(columns=['Province/State','Lat', 'Long'])
df = df.groupby('Country/Region').agg('sum')
df.head(10)
```

```python
# 將 DataFrame 作轉置，把日期變成 index
df = df.T
df.head(10)
```

```python
# 將 index 轉為 datatime 的型態
df.index = pd.to_datetime(df.index)
df.head(10)
```

## 轉為 DatetimeIndex 後就可以使用很多方便的功能

```python
df[df.index.day == 15]
```

```python
# 任何年份的 4 月
# df[df.index.month == 4]

# 2020 的 4 月
# df['2020-04']

# 2020 的 4 月
df[(df.index.month == 4) & (df.index.year == 2020)]
```

```python
df['2020-04-01':'2020-04-05']
```

```python
# 根據資料最後一天的確診人數來排序國家，稍微看一下目前比較嚴重的地區有哪些
df = df.sort_values(by=df.index[-1], axis=1,ascending=False)
df = df.iloc[:,0:10]
df
```
