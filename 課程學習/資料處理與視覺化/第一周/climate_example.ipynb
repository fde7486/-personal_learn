# Merge

<img src="image/merge.png" width="70%">

## Merge 基礎範例

```python
df1 = pd.DataFrame({'key': ['foo', 'bar', 'baz', 'foo', "tom"], 'value': [1, 2, 3, 5, 6]})
df1
```

```python
df2 = pd.DataFrame({'key': ['foo', 'bar', 'baz', 'foo', 'cat', 'dog'], 'value': [5, 6, 7, 8, 7, 9]})
df2
```

```python
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1, df2, on='key', how ='outer')
```

```python
pd.merge(df1, df2, on='key', how ='left')
```

```python
pd.merge(df1, df2, on='key', how ='right')
```

## 對 climate_temp 及 climate_precip 做 merge

```python
inner_merged_total = pd.merge(climate_temp, climate_precip, on = ["STATION", "DATE"])
inner_merged_total
```

```python
outer_merged = pd.merge(climate_temp, climate_precip, how="outer", on=["STATION", "DATE"])
outer_merged
```

```python
left_merged = pd.merge(climate_temp, climate_precip, how="left", on=["STATION", "DATE"])
left_merged
```

# 補充
## Concat

<img src="image/concat_row.png" width="70%">

<img src="image/concat_column.png" width="70%">

```python
train_df = pd.read_csv('titanic_data/train.csv')
train_df
```

```python
test_df = pd.read_csv('titanic_data/test.csv')
test_df
```

```python
# axis 為 0 就會對 row 做 concat，axis 為 1 則會對 column 做 concat
concatenated = pd.concat([train_df, test_df], axis = 0)
concatenated
```

```python
# 對 DataFrame 做操作的時候，index 不會自動更新，需要自己呼叫 reset_index
concatenated.reset_index(drop = True)
```
